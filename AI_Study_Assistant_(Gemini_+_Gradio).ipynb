##Important necessary python libraries

```python
import os
from getpass import getpass
import gradio as gr
from google import genai
```

##  Enter API Key securely

```python
os.environ["GEMINI_API_KEY"] = getpass("Enter your Gemini API Key: ")
```

##  Create client

```python
client = genai.Client(api_key=os.environ["GEMINI_API_KEY"])
```

## AI function

```python
def study_assistant(topic):
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=f"""
        You are a helpful AI study assistant.
        Explain the topic: {topic} in simple terms.
        Then generate 3 important interview questions with answers.
        Keep it structured and clear.
        """
    )

    return response.text
```

## Gradio UI

```python
interface = gr.Interface(
    fn=study_assistant,
    inputs=gr.Textbox(label="Enter Study Topic"),
    outputs=gr.Markdown(label="AI Response"),
    title=" AI Study Assistant (Gemini + Gradio)",
    description="Enter any topic and get explanation + interview questions."
)

interface.launch(share=True)
```
